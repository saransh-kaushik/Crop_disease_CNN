```python
t_model = tf.keras.models.load_model('trained_model.keras')
```

```python
import cv2 ## opencv
```

```python
## /root/.cache/kagglehub/datasets/vipoooool/new-plant-diseases-dataset/versions/2/test/test ~~~~ test data path

image_path ="/content/CornCommonRust1.JPG"
img = cv2.imread(image_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.xticks([])
plt.yticks([])
plt.title('test img')
plt.show()
```

```python
image = tf.keras.preprocessing.image.load_img(image_path, target_size=(128,128))
input_arr = tf.keras.preprocessing.image.img_to_array(image)
print(input_arr.shape)
input_arr = np.array([input_arr])
print(input_arr.shape)
```

```python
predictions = t_model.predict(input_arr)
predictions,predictions.shape
```

```python
result_index = np.argmax(predictions)
result_index
```
